matplotlib used for data visualisation

legend always in tuple

```python
import matplotlib.pyplot as plt
```

```python
abhi_marks = [34, 56, 78, 45, 54, 88]
rahul_marks = [43, 89, 98, 56, 54, 50]
plt.plot([1, 2, 3, 4, 5, 6], abhi_marks)
plt.plot([1, 2, 3, 4, 5, 6], rahul_marks)
plt.title('Marks')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid()
for x, y in enumerate(abhi_marks, 1):
    plt.text(x, y, str(y))

for x, y in enumerate(rahul_marks, 1):
    plt.text(x, y, str(y))


plt.axhline(y=50, color='red')
plt.legend(('Abhishek', 'Rahul'))
plt.xticks(range(1, 7), ['English', 'Maths', 'Computer', 'DBMS', 'AI', 'DLD'])
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x_data = np.linspace(0, 2*np.pi, 100)
y_data1 = np.sin(x_data)
y_data2 = np.cos(x_data)

plt.subplot(221)
plt.plot(x_data, y_data1)

plt.subplot(222)
plt.plot(x_data, y_data2)

plt.subplot(223)
plt.plot(x_data, y_data1, color='red')

plt.subplot(224)
plt.plot(x_data, y_data2, color='yellow')

plt.show()
```
