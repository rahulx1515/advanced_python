Question 1: Signal Smoothing using Moving Average
Scenario:
You're analyzing noisy sensor data.
Problem Statement:
Given a 1D NumPy array representing signal values, plot both the original and smoothed signal using a moving average
filter of window size k.
Input Format:
- signal: NumPy 1D array
- k: Integer (window size)
Output Format:
- 2D line plot: original (blue) and smoothed (red)
Sample Input:
signal = np.array([3, 5, 2, 8, 10, 6, 4])
k = 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal = np.array([3, 5, 2, 8, 10, 6, 4])
k = 3
def moving_average(signal, k):
    smoothed = np.convolve(signal, np.ones(k)/k, mode='valid')
    return smoothed

smoothed_signal = moving_average(signal, k)

x_original = np.arange(len(signal))
x_smoothed = np.arange(k-1, len(signal))


plt.figure(figsize=(8, 4))
plt.plot(x_original, signal, label='Original Signal', color = 'blue', marker='o')
plt.plot(x_smoothed, smoothed_signal, label='Smoothed Signal (moving avg)', color='red', marker='o')
plt.title('Signal something using moving average')
plt.xlabel('Index')
plt.ylabel('Signal Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
